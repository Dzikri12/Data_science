### scott's rule for determine bins
use to determining best bins for make a histogram

```python
data = df['viewCount']

# Calculate the number of bins using Scott's Rule
num_bins = int((max(data) - min(data)) / (3.5 * np.std(data) / len(data)**(1/3)))
num_bins
```

# data preparation

```python
df_coaster=df_coaster[['coaster_name', 
    #'Length', 'Speed', 
    'Location', 'Status', 
    #'Opening date',
      # 'Type', 
       'Manufacturer', 
    #'Height restriction', 'Model', 'Height',
       #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
       #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
      # 'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       #'Track layout', 'Fastrack available', 'Soft opening date.1',
       #'Closing date', 
    'Opened', #'Replaced by', 'Website',
       #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
      # 'Single rider line available', 'Restraint Style',
      # 'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced','latitude', 'longitude', 'Type_Main',
      'opening_date_clean',
    #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', #'height_value', 'height_unit', 
    'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df_coaster.shape
```

```python
df_coaster['opening_date_clean']=pd.to_datetime(df_coaster['opening_date_clean'])
```

```python
df_coaster
```

```python
# rename columns
df_coaster.columns
```

```python
df_coaster=df_coaster.rename(columns={'coaster_name':'Coaster_Name',
                   'year_introduced':'Year_Introduced',
                   'opening_date_clean':'Opening_Date',
                   'speed_mph':'Speed_mph', 
                   'height_ft':'Height_ft', 
                   'Inversions_clean':'Inversions',
                   'Gforce_clean':'Gforce'
                  })
```

```python
df_coaster.isna().sum()
```

```python
df_coaster.loc[df_coaster.duplicated(subset=['Coaster_Name'])]
```

```python
#checking example duplicaes
df_coaster.query('Coaster_Name=="Derby Racer"')
```

```python
df_coaster=df_coaster.loc[~df_coaster.duplicated(
    subset=['Coaster_Name','Location',
            'Opening_Date'])].reset_index(drop=True).copy()
```

```python
df_coaster.shape
```

# Feature Understanding

```python
ax=df_coaster['Year_Introduced'].value_counts().head(10).plot(kind='bar',
                                                   title='Top 10 Years Coasters Introduced')
ax.set_xlabel('Year Introduced')
ax.set_label('Counts')
```

```python
ax=df_coaster['Speed_mph'].plot(kind='hist',
                     bins=20,
                     title='Coaster Speed (mph) Distribution')
ax.set_xlabel('Speed (mph)')
ax.set_ylabel('Freq')
```

```python
ax=df_coaster['Speed_mph'].plot(kind='kde',
                     title='Coaster Speed (mph) Distribution')
ax.set_xlabel('Speed (mph)')
ax.set_ylabel('Freq')
```

# Feature Relationships

```python
df_coaster.plot(kind='scatter',
        x='Speed_mph',
        y='Height_ft',
        title='Coaster speed vs. height')
plt.show()
```

```python
sns.scatterplot(x='Speed_mph',y='Height_ft',hue='Year_Introduced',data=df_coaster)
```

```python
sns.pairplot(df_coaster,vars=['Year_Introduced','Speed_mph',
                      'Height_ft','Inversions','Gforce'],
             hue='Type_Main')
plt.show()
```

```python
df_corr=df_coaster[['Year_Introduced','Speed_mph',
                      'Height_ft','Inversions','Gforce']].dropna().corr()
df_corr
```
